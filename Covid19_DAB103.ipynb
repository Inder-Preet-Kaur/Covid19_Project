```python
#which country has highest vaccinated people
people_fully_vaccinated=dataset1['people_fully_vaccinated']
location=dataset1['location']
data1={'location':location,
      'people_fully_vaccinated':people_fully_vaccinated
    
}
people=pd.DataFrame(data1)
new=people.dropna()
new.head(30)
full=pd.DataFrame(new.groupby('location')['people_fully_vaccinated'].sum())
full.head()
full.sort_values(by=['people_fully_vaccinated'],inplace=True,ascending=False)
full
cols2 = ['location', 'people_fully_vaccinated']
full=full[:30].reset_index()
full
```

```python
plt.figure(figsize=(16, 7))
plt.bar(full.location,full.people_fully_vaccinated, color = 'c')

plt.title('Total people fully-vaccinated per country')
plt.xticks(rotation = 90)
plt.ylabel('Number of fully-vaccinated citizens (per 10 Million)')
plt.xlabel('Countries')
plt.show();
```

```python
people_vaccinated=dataset1['people_vaccinated']
location=dataset1['location']
data1={'location':location,
      'people_vaccinated':people_vaccinated
    
}
people1=pd.DataFrame(data1)
new1=people1.dropna()
new1.head(30)
partial=pd.DataFrame(new1.groupby('location')['people_vaccinated'].sum())
partial.head()
partial.sort_values(by=['people_vaccinated'],inplace=True,ascending=False)
partial
cols3 = ['location', 'people_vaccinated']
partial=partial[:30].reset_index()
partial
```

```python
plt.figure(figsize=(16, 7))
plt.bar(partial.location,partial.people_vaccinated, color = 'c')

plt.title('Total people partially-vaccinated per country')
plt.xticks(rotation = 90)
plt.ylabel('Number of partially-vaccinated citizens (per 10 Million)')
plt.xlabel('Countries')
plt.show();
```
